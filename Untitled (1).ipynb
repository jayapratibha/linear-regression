```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# for x take all the rows and column 1 
#ie.. level as matrix not vector and y is salary as vector
# we are hr and we are gonna hire negotiate future salary so we take prev salaries
# and position levels so we can detect false salaries
```

```python
y
```

```python
# not  splitting the dataset into testing and training because the data is little
```

```python
#fitting linear regression to dataset
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# fitting polunomial regression to datset by including x degrees 
#and fitting and transforming x
# polyreg automatically adds constant of 1's
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
X_poly
```

```python
#visualizing the linear regression drawa a line from dot to blue line and extend it to salary and decide
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#visualizing the polynomial regression
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
lin_reg.predict([[6.5]])
```

```python
# Predicting a new result with Polynomial Regression
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
